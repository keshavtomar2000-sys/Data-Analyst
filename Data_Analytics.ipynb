4. What is Prompt Engineering, and explain any two types of prompts used in Generative AI?

- Prompt Engineering is the process of designing and crafting effective prompts (inputs or instructions) to guide a Generative AI model (like GPT, Claude, or Gemini) to produce the best, most relevant, and accurate output.
It’s the art and science of talking to AI in the right way to get useful results.

(I) Instruction-Based Prompts-

These prompts give clear directions or tasks to the AI about what to do.

(II) Few-Shot Prompts-

These prompts include a few examples of the desired input–output pattern before asking the model to produce a new one.

8. Explain the evolution of analytics, highlighting the different stages from Descriptive to Generative AI.

- Analytics has evolved through five key stages, each building on the previous one.
These stages move from understanding the past to predicting and creating the future.

(i) Descriptive Analytics — “What Happened?”

To describe and summarize historical data — understand what has already happened.
Uses reports, dashboards, and KPIs to show trends, patterns, and summaries.

(ii)- Diagnostic Analytics — “Why Did It Happen?”

To explain why things happened by digging into relationships and root causes.
Analyzes data correlations and patterns to uncover reasons behind outcomes.
Uses statistical analysis and data mining techniques.

(iii)- Predictive Analytics — “What Will Happen?”

To forecast future outcomes based on historical data using machine learning models. Uses regression, classification, and forecasting models to make predictions.

(iv)- Prescriptive Analytics — “What Should We Do?”

To recommend actions that will produce the best outcomes. Combines predictive models with optimization algorithms and simulation.
Suggests data-driven decisions and automates responses.

(v)- Generative AI Analytics — “What Can We Create or Automate?”

To generate new insights, content, or solutions automatically using Large Language Models (LLMs) and Generative AI.
